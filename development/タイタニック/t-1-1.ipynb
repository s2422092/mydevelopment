```python
df.dropna(inplace=True)
```

```python
df= df.replace({'Sex' :{'male':0,'female':1}})
df= df.replace({'Embarked' :{'S':0,'C':1,'Q':2}})
df.head()
```

```python
X = df.drop(columns=['Survived', 'Name', 'Ticket','Cabin','Ticket','Embarked','Age'])
y=df['Survived']
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y)
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier()
clf.fit(X_train,y_train)

#正答率を求める
print("正答率(学習) = ", clf.score(X_train,y_train))
print("正答率(テスト) = ", clf.score(X_test,y_test))
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1.0)##cの値をうまく調整することによってやる。
clf.fit(X_train,y_train)
print("正答率(訓練) = ",clf.score(X_train,y_train))
print("正解(テスト)", clf.score(X_test,y_test))
```

```python
from sklearn import linear_model
clf=linear_model.SGDClassifier(loss='hinge',penalty='l2',alpha=0.001)
clf.fit(X_train,y_train)
print("正答率(訓練) = ",clf.score(X_train,y_train))
print("正解(テスト)", clf.score(X_test,y_test))
```
