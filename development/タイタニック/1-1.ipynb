```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mglearn
from IPython.display import display
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.datasets import fetch_openml
```

# データの読み込み

```python
#データの読み込み
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('titanic_test.csv')
display(df_train.head())
```

```python
print(df_train.shape)
print(df_test.shape)
```

# データの簡単な概要を表示

```python
df_train.info(), df_test.info()
```

```python
df_train.sample(20)
```

```python
df_train.drop(columns=['Cabin'],inplace=True)
df_test.drop(columns=['Cabin'],inplace=True)
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train['Embarked'].fillna('S',inplace=True)
```

```python
df_test['Fare'].fillna(df_test['Fare'].mean(), inplace=True)
```

```python
df=pd.concat([df_train,df_test],sort=True).reset_index(drop=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.corr(numeric_only=True)['Age'].abs()
```

```python
df_Age_mean=df.groupby(['Sex', 'Pclass']).median(numeric_only=True)['Age']
df_Age_mean
```

```python
df['Age']=df.groupby(['Sex','Pclass'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df['Title']=df['Name'].str.split(", ",expand=True)[1].str.split(".",expand=True)[0]
```

```python
df['Title'].value_counts()
```

```python
df['Title'] = df['Title'].replace(['Lady', 'the Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
```

```python
df['Title'].value_counts()
```

```python
df['Family_size']=df['SibSp'] + df['Parch'] + 1
```

```python
df.sample(10)
```

```python
df.drop(columns=['Name','Parch','SibSp','Ticket'],inplace=True)
```

```python
df.sample(10)
```

```python
def family_size(number):
    if number==1:
        return "Alone"
    elif number>1 and number <5:
        return "Small"
    else:
        return "Large"
```

```python
df['Family_size']=df['Family_size'].apply(family_size)
```

#  データ分析

```python
df.info()
```

```python
df['Age'] = df['Age'].astype('int64')
```

```python
df.info()
```

```python
col=['Embarked','Pclass','Sex','Title','Family_size']
for i in col:
    print('Survival Correlation by:', i)
    df2=df.groupby(i)['Survived'].mean().reset_index()
    print(df2)
    print('-'*10, '\n')
```

```python
import plotly.express as px
import plotly.graph_objects as go


fig = plt()

fig.add_trace(go.Histogram(x=df[df['Survived']==0]['Age'], name='Not Survived', opacity=0.5))
fig.add_trace(go.Histogram(x=df[df['Survived']==1]['Age'], name='Survived', opacity=0.5))

fig.update_layout(
    title='Age Distribution by Survival',
    xaxis_title='Age',
    yaxis_title='Density',
    barmode='overlay',  
    bargap=0.1, 
)

fig.show()
```
