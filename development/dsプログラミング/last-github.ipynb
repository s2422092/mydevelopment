```python
import requests
from bs4 import BeautifulSoup

# URL（例）
href = 'https://github.com/orgs/google/repositories'

# URLからページ内容を取得
response = requests.get(href)

# レスポンスの内容を解析
soup = BeautifulSoup(response.text, 'html.parser')

# リポジトリ情報が入っている <li> タグを取得
li_tags = soup.find_all('li', id=':R5ab:-list-view-node-:R3tab:')

# 各リポジトリ情報を取得
for li in li_tags:
    # リポジトリ名を取得
    repo_name_span = li.find('span', class_='prc-Text-Text-0ima0')
    repo_name = repo_name_span.get_text(strip=True) if repo_name_span else "なし"
    
    # 主要言語を取得
    language_span = li.find('span', class_='Box-sc-g0xbh4-0 fVplbS prc-Text-Text-0ima0')
    language = language_span.get_text(strip=True) if language_span else "なし"
    
    # スター数を取得
    star_span = li.find('span', class_='prc-Text-Text-0ima0')
    star_count = "なし"
    
    # aria-label属性に 'stars' が含まれているかを確認してスター数を取得
    if star_span:
        aria_label = star_span.get('aria-label')
        if aria_label and 'stars' in aria_label.lower():
            star_count = star_span.get_text(strip=True) or "なし"
    
    # 結果を出力
    print(f"リポジトリ名: {repo_name}, 主要言語: {language}, スター数: {star_count}")
```

## スター数を取得

```python
import requests
from bs4 import BeautifulSoup

# URLを設定（適宜URLを変更）
href = "https://github.com/orgs/google/repositories?page=1"

# URLからページ内容を取得
response = requests.get(href)

# レスポンスの内容を解析
soup = BeautifulSoup(response.text, 'html.parser')

# aria-label 属性を持つ <a> タグを検索
a_tags = soup.find_all('a', attrs={'aria-label': True})

# カウントを初期化
count = 0

# 数字を変換する関数を定義
def convert_to_number(label):
    # "k", "M", "B" などの単位を処理
    if 'k' in label:
        return float(label.replace('k', '').strip()) * 1000
    elif 'M' in label:
        return float(label.replace('M', '').strip()) * 1000000
    elif 'B' in label:
        return float(label.replace('B', '').strip()) * 1000000000
    else:
        # 単位がない場合、そのまま数値として返す
        return float(label.strip())

# 見つかった <a> タグ内で aria-label の内容を出力
for a in a_tags:
    aria_label = a.get('aria-label')  # <a> タグの 'aria-label' 属性を取得

    # <a> タグ内に <svg> タグが存在するか、かつ <svg> タグ内に <a> タグの href 属性が '/stargazers' で終わるものをチェック
    svg_tag = a.find('svg')  # <a> タグ内の <svg> タグを検索
    if svg_tag:
        # <svg> タグ内にリンク（<a>タグ）があれば、その href 属性をチェック
        href_tag = svg_tag.find_parent('a', href=True)  # svg タグが親の <a> タグに紐づいているかを調べる
        if href_tag and href_tag.get('href', '').endswith('/stargazers'):
            if aria_label:
                # "stars" を削除し、単位変換を行う
                aria_label = aria_label.replace("stars", "").strip()  # "stars" を削除
                number = convert_to_number(aria_label)  # 数値に変換
                print(f"{number:.0f}")  # 数値を整数で表示
                count += 1  # カウントを増加

# カウントを表示
print(f"\naria-label 属性を持つ <a> タグのうち、<svg> タグ内に /stargazers を含むリンクの個数: {count}")


```

## リンク先におけるリポジトリ名の取得全てのリポジトリ名を取得する

```python
import requests
from bs4 import BeautifulSoup
import time

# 基本のURLを設定
base_url = "https://github.com/orgs/google/repositories"

# リストにリンクをまとめる
href_list = []

# 1ページ目から90ページ目まで繰り返し
for i in range(1, 91):
    href = base_url + "?page=" + str(i)  # 基本URLにページ番号を追加
    href_list.append(href)  # リストにURLを追加
    time.sleep(1)  # 1秒待機

# 各ページのURLで内容を取得し、必要な情報を抽出
total_count = 0  # <span> タグの全体のカウント

# リストのリンクを1つずつ処理
for href in href_list:
    # URLからページ内容を取得
    response = requests.get(href)
    
    # レスポンスの内容を解析
    soup = BeautifulSoup(response.text, 'html.parser')

    # class="Box-sc-g0xbh4-0 cCxlBe" の <div> タグを検索
    div_tags = soup.find_all('div', class_='Box-sc-g0xbh4-0 cCxlBe')

    page_count = 0  # 各ページで見つかった <span> タグのカウント
    spans_found = set()  # 見つかった <span> タグの内容を重複なしで保持するためのセット

    # <div> タグ内の <a> タグ内の <span> タグを検索
    for div in div_tags:
        # div タグ内のさらに別の div タグ（class="list-view-item-title-container"）を検索
        inner_div_tags = div.find_all('div', class_='Box-sc-g0xbh4-0 jehEGV Title-module__container--l9xi7')

        # 各 inner div タグ内の <a> タグ内の <span> タグを検索
        for inner_div in inner_div_tags:
            a_tags = inner_div.find_all('a')  # <div> タグ内の <a> タグを検索

            # 各 <a> タグ内の <span> タグを検索
            for a in a_tags:
                span_tags = a.find_all('span')  # <a> タグ内の <span> タグを検索

                # <span> タグが存在する場合、そのテキストを出力
                for span in span_tags:
                    span_text = span.get_text().strip()  # <span> タグのテキストを取得し、余分な空白を削除
                    if span_text not in spans_found:  # 重複しない場合のみ処理
                        print(f"ページ {href} の <span> タグの内容: {span_text}")  # <span> タグのテキストを表示
                        spans_found.add(span_text)  # 見つかった <span> タグの内容をセットに追加
                        page_count += 1  # ページごとのカウントを増加させる

    # class="Box-sc-g0xbh4-0 fCvgBf" の <div> タグを検索
    div_tag = soup.find_all('div', class_='Box-sc-g0xbh4-0 fCvgBf')

    # 見つかった <div> タグ内の <span> タグを検索
    for div in div_tag:
        # この <div> タグ内にあるすべての <span> タグを検索
        span_tags = div.find_all('span')
        
        for span in span_tags:
            span_text = span.get_text().strip()  # <span> タグのテキストを取得し、余分な空白を削除
            if span_text not in spans_found:  # 重複しない場合のみ処理
                print(f"ページ {href} の 追加の <span> タグの内容: {span_text}")  # 追加の <span> タグ内のテキスト
                spans_found.add(span_text)  # 見つかった <span> タグの内容をセットに追加
                total_count += 1  # 合計カウントを増加

    # 各ページのカウントを表示
    print(f"ページ {href} で {page_count} 個の <span> タグが見つかりました。")
    total_count += page_count  # 合計カウントを増加させる

    time.sleep(1)  # 各ページごとに1秒待機

# 最終的な合計カウントを表示
print(f"合計 {total_count} 個の <span> タグが見つかりました。")
```

```python
import requests
from bs4 import BeautifulSoup
import time

# リポジトリ情報を取得する処理を関数にまとめる
def get_info_from_page(page_url):
    # URLからページ内容を取得
    response = requests.get(page_url)
    
    # レスポンスの内容を解析
    soup = BeautifulSoup(response.text, 'html.parser')

    # <a> タグをすべて取得（aria-labelがなくても含める）
    a_tags = soup.find_all('a')

    # カウントを初期化
    count = 0

    # 見つかった <a> タグ内で aria-label の内容やその他の情報を出力
    for a in a_tags:
        aria_label = a.get('aria-label')  # 'aria-label' 属性を取得
        href = a.get('href')  # 'href' 属性を取得

        # aria-labelが存在する場合、それを出力
        if aria_label:
            print(f"aria-label: {aria_label}")

        # hrefが存在する場合、それを出力
        if href:
            print(f"リンク: {href}")

        # 他にもリンクに関連する情報を出力することができます。
        count += 1  # 出力されたリンクの数をカウント

    return count

# 基本のURL（ページ番号を後で追加）
base_url = "https://github.com/orgs/google/repositories?page={}"

# 全ページを処理
total_count = 0
for i in range(1, 91):  # ページ番号1から90まで
    page_url = base_url.format(i)  # ページ番号を更新してURLを作成
    print(f"\n処理中のページ: {page_url}")
    
    # 各ページで情報を取得
    count = get_info_from_page(page_url)
    total_count += count
    
    # 1秒待機して、サーバーへの負荷を避ける
    time.sleep(1)

# 全ページで取得したリンク数の合計を表示
print(f"\n全ページで見つかったリンクの合計数: {total_count}")
```
