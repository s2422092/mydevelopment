##  スターの数を取得する

```python
import requests
from bs4 import BeautifulSoup

# GitHubのURLテンプレート
base_url = "https://github.com/orgs/google/repositories?page={}"

# 出力された行数をカウントする変数
line_count = 0

# 1ページから90ページまで繰り返し処理
for page_num in range(1, 91):
    # URLをページ番号でフォーマット
    url = base_url.format(page_num)

    # ページの内容を取得
    response = requests.get(url)

    # レスポンスが正常でない場合はスキップ
    if response.status_code != 200:
        print(f"ページ {page_num} の読み込みに失敗しました。")
        continue

    # レスポンスの内容を解析
    soup = BeautifulSoup(response.text, 'html.parser')

    # aria-label 属性を持ち、かつその中に "stars" が含まれている <a> タグを検索
    a_tags = soup.find_all('a', attrs={'aria-label': True})

    # 各ページのリポジトリとスター数を取得
    for a in a_tags:
        aria_label = a.get('aria-label')  # aria-label 属性を取得
        
        if 'star' in aria_label.lower():  # aria-label に "stars" が含まれているか確認
            # <a> タグ内のテキストを取得（数字が含まれている場合）
            text = a.get_text(strip=True)
            
            if text.isdigit():  # テキストが数字だけの場合
                print(f"ページ {page_num}: スター数 {text}")
                line_count += 1  # 出力行数をカウント

            else:
                print("記入なし")
                line_count += 1

# 最終的な出力行数を表示
print(f"\n出力された行数: {line_count}")
```

```python
import requests
from bs4 import BeautifulSoup
import time

# 基本のURL（ページ番号を後で追加）
href = "https://github.com/orgs/google/repositories?page=1"

# リポジトリ情報を取得する処理を関数にまとめる
def get_repositories(page_url, start_index):
    # URLからページ内容を取得
    response = requests.get(page_url)
    
    # レスポンスの内容を解析
    soup = BeautifulSoup(response.text, 'html.parser')

    # class="Box-sc-g0xbh4-0 fCvgBf" の <div> タグを検索
    div_tags = soup.find_all('div', class_='Box-sc-g0xbh4-0 fCvgBf')

    # カウントを初期化
    count = 0

    # 指定したクラスの <div> タグが見つからなかった場合は「欠損値」と出力
    if div_tags:
        for div in div_tags:
            # この <div> タグ内にあるすべての <span> タグを検索
            span_tags = div.find_all('span')
            
            if span_tags:
                for span in span_tags:
                    # 番号をつけてテキストを表示
                    print(f"{start_index}. {span.get_text()}")  # 番号付きで <span> タグ内のテキストを表示
                    count += 1  # カウントを増加
                    start_index += 1  # 番号をインクリメント
            else:
                print(f"{start_index}. この <div> タグ内には <span> タグが見つかりませんでした。")
                start_index += 1
    else:
        print(f"{start_index}. 指定されたクラスの <div> タグは見つかりませんでした。")
        print(f"{start_index}. 欠損値：該当する <div> タグが見つからなかったため、データが欠損しています。")
        count = None  # 欠損値としてNoneを設定する
        start_index += 1

    return count, start_index

# 1ページ目から90ページ目まで処理
total_count = 0
start_index = 1  # 番号の初期値

for i in range(1, 91):
    page_url = href.replace('page=1', f'page={i}')  # ページ番号を更新
    print(f"\n処理中のページ: {page_url}")
    
    # 各ページでリポジトリ情報を取得
    count, start_index = get_repositories(page_url, start_index)
    if count is not None:  # 欠損値でない場合のみ合計に加算
        total_count += count
        
    else:
                print("記入なし")
                line_count += 1
    
    # 1秒待機して、サーバーへの負荷を避ける
    time.sleep(1)

# 全ページで見つかった <span> タグの合計数を表示
print(f"\n全ページで見つかった <span> タグの合計数: {total_count}")


```
