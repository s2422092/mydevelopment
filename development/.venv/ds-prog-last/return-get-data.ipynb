```python
import requests
from bs4 import BeautifulSoup

# WebページのURL
main_url = 'https://www.jalan.net/'

# Webページを取得
response = requests.get(main_url)
response.encoding = response.apparent_encoding  # エンコーディングを適切に設定

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, 'html.parser')

# id="travelNavOnSiteHeader"を持つ<li>タグを取得
li_tag = soup.find('li', {'id': 'travelNavOnSiteHeader'})

# <ul>タグ内のすべての<a>タグを取得
if li_tag:
    a_tags = li_tag.find_all('a')
    for a_tag in a_tags:
        # <a>タグのhrefとテキストを出力
        href = a_tag.get('href')
        text = a_tag.text.strip()  # テキスト部分を取得し、前後の空白を削除
        print(f"Text: {text}")
        print(f"Href: {href}")
```

```python
import requests
from bs4 import BeautifulSoup

# WebページのURL
main_url = 'https://www.jalan.net/kankou/?ccnt=global_navi'

# Webページを取得
response = requests.get(main_url)
response.encoding = response.apparent_encoding  # エンコーディングを適切に設定

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, 'html.parser')

# <div id="mapArea" class="has-han">内の<dd>タグと<dt>タグを検索
map_area = soup.find('div', {'id': 'mapArea'})  # id="mapArea"を持つdivを検索
dd_tags = map_area.find_all('dd')  # <dd>タグを取得
dt_tags = map_area.find_all('dt')  # <dt>タグを取得（沖縄など）


# 取得した<dd>タグ内の<a>タグのhrefを出力
for dd_tag in dd_tags:
    a_tag = dd_tag.find('a')  # <dd>タグ内の<a>タグを検索
    if a_tag:
        href = a_tag.get('href')  # href属性を取得
        if href:  # hrefが存在する場合のみ出力
            print(f"Found href: {'https://www.jalan.net' + href}")


# <div class="okinawa">内の<a>タグのhrefを出力
okinawa_div = soup.find('div', {'class': 'okinawa'})  # class="okinawa"のdivを検索
if okinawa_div:
    a_tag = okinawa_div.find('a')  # <div class="okinawa">内の<a>タグを検索
    if a_tag:
        href = a_tag.get('href')  # href属性を取得
        if href:  # hrefが存在する場合のみ出力
            print(f"Found href: {'https://www.jalan.net' + href}")

```

```python
import requests
from bs4 import BeautifulSoup
import time

# URLを設定
href = 'https://www.jalan.net/kankou/120000/'

# Webページを取得
response = requests.get(href)
response.encoding = response.apparent_encoding  # エンコーディングを適切に設定

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, 'html.parser')

# divタグを取得
div_tags = soup.find('div', {'class': 'rankList'}) 

# 1秒間スリープ（待機）
time.sleep(1)

# 順番に番号とh3の内容を1行で出力
if div_tags:
    for index, h3_tag in enumerate(div_tags.find_all('h3'), start=1):
        print(f"{index}、{h3_tag.text.strip()}")
```
