##  それぞれのここの観光地ページの前のページからデータを取得できるようにした。

```python
import requests
from bs4 import BeautifulSoup
import time
import re  # 正規表現を使用するためインポート

# URLを設定
href = 'https://www.jalan.net/kankou/120000/'

# Webページを取得
response = requests.get(href)
response.encoding = response.apparent_encoding  # エンコーディングを適切に設定

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.text, 'html.parser')

# divタグを取得（rankList内）
div_tags = soup.find('div', {'class': 'rankList'}) 

# 1秒間スリープ（待機）
time.sleep(1)

# divタグが存在する場合
if div_tags:
    # divタグ内のすべてのh3タグを取得
    for index, h3_tag in enumerate(div_tags.find_all('h3'), start=1):
        name = h3_tag.text.strip()  # 名前（h3タグ内のテキスト）

        # レビュー評価と口コミ数を取得する
        rating_div = h3_tag.find_next('div', class_=['rating', 'ratingL'])  # h3タグの次にある評価関連のdiv
        review_point = None
        review_count = None

        if rating_div:
            # レビュー評価（例：4.6）を取得
            review_point_tag = rating_div.find('span', class_='reviewPoint')
            if review_point_tag:
                review_point = review_point_tag.text.strip()

            # 口コミ数（例：14,994件）を取得
            review_count_tag = rating_div.find('span', class_='reviewCount')
            if review_count_tag:
                review_count_text = review_count_tag.text.strip()
                # 口コミ件数に含まれる数字を抽出
                match = re.search(r'(\d{1,3}(?:,\d{3})*)', review_count_text)
                if match:
                    review_count = match.group(1)

        # 出力形式: 番号、名前、レビュー評価、口コミ数
        print(f"{index}、{name}、{review_point if review_point else 'N/A'}、{review_count if review_count else 'N/A'}")
```
