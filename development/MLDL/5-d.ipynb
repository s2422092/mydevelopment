```python
# DBSCANで外れ値を得る

from sklearn import datasets
import matplotlib.pyplot as plt

X, y = datasets.make_blobs(n_samples=200, centers=1, n_features=2, cluster_std=2, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
```

```python
# DBSCAN
from sklearn import cluster
cls = cluster.DBSCAN(eps=1,min_samples=5)
cls.fit(X)
y=cls.labels_

# 図を表示
plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], c=y)
```
