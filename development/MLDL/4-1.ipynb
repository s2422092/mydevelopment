```python
# 回帰分析に適したサンプルデータの作成（覚えなくてよい）
import mglearn
X, y = mglearn.datasets.make_wave(n_samples=100)

# 作成したサンプルデータを表示
import matplotlib.pyplot as plt
plt.plot(X, y, 'o')
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# 線形モデルのインポート
from sklearn import linear_model

# SGD回帰
rg = linear_model.SGDRegressor()

#モデルの適用
rg.fit(X,y)

# 偏回帰係数の表示
print("coefficient  = ", rg.coef_)
# 切片の表示
print("intercept = ", rg.intercept_)
# 決定係数の表示
print(rg.score(X,y))

# グラフ表示
plt.plot(X, y, 'o')
plt.plot(X, rg.predict(X), color = 'red') # 回帰直線をプロット
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# SVMのインポート
from sklearn import svm
# SVR(rbfカーネル)
rg = svm.SVR(kernel='rbf', C=1, gamma=1.1, epsilon=0.1)
# 予測値をy_predに入れる
y_pred = rg.fit(X,y).predict(X)
# 決定係数の表示
print(rg.score(X,y))


# グラフ表示
plt.scatter(X, y)
plt.plot(X, y_pred, 'o', color='red')
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# 外挿の実験
# X<1のみ学習用データとして切り出す
import numpy as np
import pandas as pd
all_data = pd.DataFrame(np.hstack([X, y.reshape(-1, 1)]))
train_data = all_data[all_data.iloc[:,0] < 1]
X_train = train_data.iloc[:,0:1]
y_train = train_data.iloc[:,1]

# X<1のデータをを学習し、全データを予測
y_pred_out = rg.fit(X_train, y_train).predict(X)

# グラフ表示
plt.scatter(X, y)
plt.plot(X, y_pred_out, 'o', color='red')
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# 決定木をインポート
from sklearn import tree
rg = tree.DecisionTreeRegressor(max_depth=5)
#予測値を入れる
y_pred = rg.fit(X, y).predict(X)
# 決定係数の表示
print(rg.score(X,y))

# グラフ表示
plt.scatter(X, y)
plt.plot(X, y_pred, 'o', color='red')
plt.xlabel("feature")
plt.ylabel("target")
```

```python
# X<1のデータをを学習し、全データを予測
y_pred_out = rg.fit(X_train, y_train).predict(X)

# グラフ表示
plt.scatter(X, y)
plt.plot(X, y_pred_out, 'o', color='red')
plt.xlabel("feature")
plt.ylabel("target")
```
