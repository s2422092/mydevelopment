```python
%matplotlib inline
from sklearn.datasets import make_blobs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
X,y = make_blobs(centers=4, random_state=8)
y=y%2

# Display Scatter Plot
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# 3D
# Library for 3D graph
from mpl_toolkits.mplot3d import Axes3D

x1 = X[:,0]
z1 = X[:,1]
# 
# Add nonlinear features on the vertical axis
y1 = z1 ** 2

# Draw Scatter plot in 3D graph
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter3D(np.ravel(x1), np.ravel(z1), np.ravel(y1), c=y)
plt.show()
```

```python
# import libraries
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions

# Make train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# SVC
# Apply SVC with kernel approximation
#clf = SVC(kernel='rbf')
clf = SVC(kernel='rbf', C=100, gamma=0.01)
clf.fit(X_train,y_train)

# draw scatter plot
plot_decision_regions(X, y, clf=clf)
```
